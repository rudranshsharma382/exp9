```python

# Declration of Numpy
import numpy as np

a = np.array([1 , 10 , 20 , 30 , 40 , 50 , 500 ])
b = np.array([[1,2,3],[4,5,6]])
print(a)
print(b)
```

```python
# Dmension of Array
print(a.ndim)
print(b.ndim)
```

```python
#To find shape

a.shape
b.shape
print(a.shape)
print(b.shape)

# To find data type

print(a.dtype)
print(b.dtype)

#In-Built function using Numpy
#zeros

print(np.zeros((2,3)))

#ones

print(np.ones((3,3)))

#eye

print(np.eye(3))

#arrange

print(np.arange(1,10,2))

#linspace

print(np.linspace(0,1,4))

#mean

print(np.mean(a))

#median

print(np.median(a))

#standard deviation

print(np.std(a))

#max value

print(np.max(a))

#min value

print(np.min(a))

#sum
print(np.sum(a))
```
